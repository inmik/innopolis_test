```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def show_shapiro_test(sample):
    print(sample)
    print(stats.shapiro(sample))
    sns.histplot(sample, kde = True)
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Titanic.csv")
```

```python
sample1 = df.dropna(subset=['age'])['age']
show_shapiro_test(sample1)
```
